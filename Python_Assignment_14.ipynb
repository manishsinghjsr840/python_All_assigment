1. What does RGBA stand for?

- - RGBA stands for red green blue alpha

```python
2. From the Pillow module, how do you get the RGBA value of any images?

from PIL import Image

Img = Image.open(r"SamplePNG.png")
RGBA_Img = Img.convert("RGBA")
Pixel_data = list(RGBA_Img.getdata())

for pixel in Pixel_data:
    print(pixel)

Img.show
print(f"Image mode is : {Img.mode}")
print(f"Image format is : {Img.format}")
print(f"Image size is : {Img.size}")
```

3. What is a box tuple, and how does it work?

- We can crop a picture with the crop() method of the image class. The method takes a box tuple that characterizes the position and size of the cropped area and returns an image object addressing the cropped picture.
- The coordinates for the box are (left, upper, right, lower).

```python
from PIL import Image

My_samle = Image.open("SamplePNG.png")
#My_samle.show()

Cropped_Sample = My_samle.crop(box=(50,50,250,200))
print(f"Original size of sample is {My_samle.size}")
print(f"Size of cropped sample is {Cropped_Sample.size}")
Cropped_Sample
```

```python
# 4. Use your image and load in notebook then, How can you find out the width and height of an Image object?

from  PIL import Image 
import matplotlib.pyplot as plt

Sample = Image.open("SamplePNG.png")
plt.imshow(Sample)
Sample

width, height = Sample.size

print("Width:", width)
print("Height:", height)
```

```python
# 5. What method would you call to get Image object for a 100×100 image, excluding the lower-left quarter of it?


from  PIL import Image

Cropped_Sample = Sample.crop(box=(0,50,50,100))
plt.imshow(Cropped_Sample)
#Cropped_Sample.show()
```

```python
# 6. After making changes to an Image object, how could you save it as an image file?

from  PIL import Image

Cropped_Sample = Sample.crop(box=(0,50,50,100))
Cropped_Sample.save("New.png")
```

7. What module contains Pillow’s shape-drawing code?

- The module that contains Pillow's shape-drawing code is the ImageDraw module

```python
# 8. Image objects do not have drawing methods. What kind of object does? How do you get this kind of object?

from  PIL import Image, ImageDraw

image = Image.open("SamplePNG.png")
draw = ImageDraw.Draw(image)

draw.rectangle([(100, 100), (400, 400)], outline="red")
draw.text((50, 50), "Akshay Gaidhani !", fill="green")

image.save("New_processed_image.png")
plt.imshow(image)
```
